# Problem 58: Regularization

```python
C = np.logspace(-10, 10, 10, base=10)
result = []

for c in C:
    lr = LogisticRegression(max_iter=100, C=c,solver='lbfgs', multi_class='auto')
    lr.fit(X_train, y_train)
    yval_pred = lr.predict(X_val)
    y_pred = lr.predict(X_test)
    
    train_acc = accuracy_score(y_train, lr.predict(X_train))
    val_acc = accuracy_score(y_val, yval_pred)
    test_acc = accuracy_score(y_test, y_pred)
    
    result.append([c, train_acc, val_acc, test_acc])
```

```python
result = np.array(result).T
plt.plot(result[0], result[1], label='train')
plt.plot(result[0], result[2], label='val')
plt.plot(result[0], result[3], label='test')

plt.legend()
plt.xscale('log')
plt.ylim(0,1.1)
plt.xlabel('C')
plt.ylabel('accuracy')
```

# Problem 59: Hyper-parameter tuning

```python
from sklearn.multiclass import OneVsRestClassifier
from sklearn.svm import SVC
from sklearn.decomposition import PCA

C = 1.
kernel = 'rbf'
gamma = 0.01

pca = PCA(n_components=10)
pca.fit(X_train)
Xtrain_pca = pca.transform(X_train)
Xtest_pca = pca.transform(X_test)

estimator = SVC(C=C, kernel=kernel, gamma=gamma)
classifier = OneVsRestClassifier(estimator)
classifier.fit(Xtrain_pca, y_train)
y_pred = classifier.predict(Xtest_pca)

print('train data accuracy: ', accuracy_score(y_train, classifier.predict(Xtrain_pca)))
print('test data accuracy: ', accuracy_score(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV

tuned_parameters = [
    {'C':[0.1, 1, 10, 100]},
    {'C':[0.1, 1, 10, 100], 'kernel':['rbf'], 'gamma':[0.001, 0.01, 0.1, 1]}
]

score = 'accuracy'
clf = GridSearchCV(SVC(gamma="scale"), tuned_parameters, cv=5, scoring=score)
clf.fit(Xtrain_pca, y_train)
```
