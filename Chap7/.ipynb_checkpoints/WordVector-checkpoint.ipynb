# Problem 64: Analogy data experiment

```python
!wget http://download.tensorflow.org/data/questions-words.txt
```

```python
with open('questions-words.txt') as f:
    questions = f.readlines()
    
with open('questions-words-add.txt', 'w') as f:
    for i, question in enumerate(questions):
        words = question.split()
        if len(words) == 4:
            ans = model.most_similar(positive=[words[1],words[2]], negative=[words[0]],topn=1)[0]
            words += [ans[0], str(ans[1])]
            output = ' '.join(words)+'\n'
        else:
            output = question
        f.write(output)
        if (i%1000==0):
            print(i)
```

# Problem 65: Accuracy score on the analogy task

```python
with open('questions-words-add.txt') as f:
    lines = f.readlines()
    
semantic, syntactic = [], []
for line in lines:
    if line.startswith(': gram'):
        ctg = 'syntactic'
    elif line.startswith(':'):
        ctg = 'semantic'
    else:
        if ctg == 'semantic':
            semantic.append(line.rstrip('\n').split())
        if ctg == 'syntactic':
            syntactic.append(line.rstrip('\n').split())

acc_sem = sum([1 if sem[3]==sem[4] else 0 for sem in semantic])/len(semantic)
acc_syn = sum([1 if syn[3]==syn[4] else 0 for syn in syntactic])/len(syntactic)

print('semantic accuracy: ', acc_sem)
print('syntactic accuracy: ', acc_syn)
```

```python
import pandas as pd

df = pd.read_csv('combined.csv')
vec_similarity = [model.similarity(v1, v2) for v1, v2 in zip(df['Word 1'], df['Word 2'])]
human_similarity = df['Human (mean)']

correlation, pvalue = spearmanr(human_similarity, vec_similarity)
print('スピアマン相関係数: ', correlation)
```

# Problem 67: k-means clustering

```python
with open('questions-words.txt') as f:
    lines = f.readlines()

cots = []
for line in lines:
    if line.startswith(': currency'):
        break
    elif line.startswith(':'):
        continue
    else:
        _, a, _, b = line.rstrip('\n').split()
        cots += [a,b]
        
country = {}
for e in set(cots):
    if e in model:
        country[e] = model[e]
```

```python
country_vec = [model[c] for c in list(country.keys())]

kmeans = KMeans(n_clusters=5)
kmeans.fit(country_vec)

for i in range(5):
    cluster = np.where(kmeans.labels_ == i)[0]
    print('cluster', i)
    print(', '.join([list(country.keys())[k] for k in cluster]))
```

# Problem 68: Ward's method clustering

```python
plt.figure(figsize=(32,24))
link = linkage(list(country.values()), method='ward')
dendrogram(link, labels=list(country.keys()), leaf_rotation=90, leaf_font_size=10)
plt.show()
```

# Problem 69: t-SNE Visualization

```python
vec_embedded = TSNE(n_components=2).fit_transform(list(country.values()))
vec_embedded_t = list(zip(*vec_embedded))
fig, ax = plt.subplots(figsize=(16,12))
plt.scatter(*vec_embedded_t)

for i,c in enumerate(list(country.keys())):
    ax.annotate(c, (vec_embedded[i][0], vec_embedded[i][1]))
```
